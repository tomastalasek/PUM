# Domácí úkol 3

Studenti, kteří byli na výuce musi vyřešit celkem čtyři z šesti domácích úkolů. Studenti, kteří na výuce absentovali musí vyřešit všechny domácí úkoly.

## Příklad 1

- Vykreslete funkci $2\cdot cos(x)$ na intervalu $\langle -\pi, \pi \rangle$. Graf funkce ať má zelenou barvu

## Příklad 2

- Vykreslete funkci $\sqrt x + 7$ na intervalu $\langle 0, 7 \rangle$ (odmocnina se počítá pomocí funkce `sqrt()`).


## Příklad 3

- Do jednoho grafu vykreslete funkce $sin(2 x)$, $2cos(x)$ a $x^2$ na intervalu $\langle -5, 5 \rangle$. Každá funkce ať je vyvedana jinou barvou a jiným stylem čáry. U druhé funkce navíc zvýrazněte vypočítané body pomocí křížku. U grafu popište obě osy, uvěďte název (Domaci ukol 3) a zobrazte legendu (pro druhou mocninu je třeba použít `.^2` - včetně tečky, jinak to nebude fungovat).
