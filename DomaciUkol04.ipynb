# Domácí úkol 4

Studenti, kteří byli na výuce musi vyřešit celkem čtyři z šesti domácích úkolů. Studenti, kteří na výuce absentovali musí vyřešit všechny domácí úkoly.

## Příklad 1

S pomocí cyklu `for` napište kód, který pro zadané $n$ vypíše součet čísel od $1$ do $n$

Příklad:
- pro n=10 program vypíše 55
- pro n=100 program vypíše 5050
- pro n=400 program vypíše 80200

## Příklad 2

S pomocí cyklu `for` výpočítejte 

$$\sum_{i=1}^{10} \frac{i^3-10}{i+1}$$


```{admonition} Zobrazit výsledek
:class: dropdown warning
 $$\sum_{i=1}^{10} \frac{i^3-10}{i+1} =317.78  $$
```
