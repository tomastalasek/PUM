# Domácí úkol 5

Studenti, kteří byli na výuce musi vyřešit celkem čtyři z šesti domácích úkolů. Studenti, kteří na výuce absentovali musí vyřešit všechny domácí úkoly.

## Příklad 1

S využitím poznatků z lekce 5 vypočítejte obsah plochy pod funkcí $\frac{\sin(x)}{x+5}+x^2+3$ na intervalu $\langle -4, 4 \rangle$.

## Příklad 2
