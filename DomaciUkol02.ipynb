## Příklad 2

Pomocí podmíněného příkazu `if` vytvořte program, který upraví zadané číslo `a` následovně
- pokud je `a` kladné, uloží do `a` jeho dvojnásobek
- pokud `a` nebude kladné, do `a` uloží číslo `a` zmenšené o 10

Tj. na prvním řádku bude uvedeno a=10 (nebo libovolné jiné číslo). Následovat bude `if`


Příklady:

- Pokud a=10, pak výsledek bude a=20. 
- Pokud se a=-5, pak výsledek bude a=-15.
